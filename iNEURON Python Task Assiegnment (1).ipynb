## Qn.1.Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word

```python
def find_highest_frequency_length(string):
    word_frequency = {}
    words = string.split()

    for word in words:
        word_frequency[word] = word_frequency.get(word, 0) + 1

    highest_frequency = max(word_frequency.values(), default=0)

    highest_frequency_word = [word for word, freq in word_frequency.items() if freq == highest_frequency]

    if highest_frequency_word:
        return len(highest_frequency_word[0])
    else:
        return 0

# Example usage:
input_string = "write write write all the number from from from 1 to 100"
result = find_highest_frequency_length(input_string)
print(result)
```

## Qn.2.Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO

```python
from collections import Counter
def is_valid_string(string):
    char_count = Counter(string)
    count_freq = Counter(char_count.values())

    if len(count_freq) == 1:
        # All characters appear the same number of times
        return "YES"

    if len(count_freq) == 2:
        # Check if removing one character can make all others occur the same number of times
        min_count, max_count = sorted(count_freq.items())

        if min_count[0] == 1 and min_count[1] == 1:
            # One character appears only once, removing it will make all others occur the same number of times
            return "YES"

        if max_count[1] == 1 and max_count[0] - min_count[0] == 1:
            # One character occurs more frequently than others, removing one occurrence of it will make all others occur the same number of times
            return "YES"

    return "NO"

# Example usage:
input_string = "abcc"
result = is_valid_string(input_string)
print(result)
```

## Qn.4.Write a program to download the data from the link given below and then read the data and convert the into
the proper structure and return it as a CSV file.
Link - https://data.nasa.gov/resource/y77d-th95.json
Note - Write code comments wherever needed for code understanding

```python
import requests
import pandas as pd

def download_data(url):
    response = requests.get(url)  # Send a GET request to the URL
    if response.status_code == 200:  # If the request is successful
        return response.json()  # Return the JSON data
    else:
        return None

def convert_to_csv(data, filename):
    if data:
        # Extract the relevant attributes from the JSON data
        processed_data = []
        for item in data:
            processed_item = {
                "Name of Earth Meteorite": item["name"],
                "id": item["id"],
                "Meteorite": item["nametype"],
                "recclass": item["recclass"],
                "mass": item["mass (g)"],
                "year": pd.to_datetime(item["year"], errors="coerce"),
                "reclat": item["reclat"],
                "reclong": item["reclong"],
                "point coordinates": item["geolocation"]["coordinates"]
            }
            processed_data.append(processed_item)

        # Create a DataFrame from the processed data
        df = pd.DataFrame(processed_data)

        # Save the DataFrame as a CSV file
        df.to_csv(filename, index=False)
        print(f"Data saved as {filename} successfully.")
    else:
        print("Failed to download data.")

if __name__ == "__main__":
    url = "https://data.nasa.gov/resource/y77d-th95.json"
    filename = "nasa_data.csv"

    nasa_data = download_data(url)  # Download data from the given URL
    convert_to_csv(nasa_data, filename)  # Convert and save the data as a CSV file
```

Insights to be drawn -
● Get all Pokemons whose spawn rate is less than 5%
● Get all Pokemons that have less than 4 weaknesses
● Get all Pokemons that have no multipliers at all
● Get all Pokemons that do not have more than 2 evolutions
● Get all Pokemons whose spawn time is less than 300 seconds.
Note - spawn time format is "05:32”, so assume “minute: second” format and perform the analysis.
● Get all Pokemon who have more than two types of capabilities

```python
import pandas as pd
import matplotlib.pyplot as plt

def analyze_data(data):
    # Create a DataFrame from the data
    df = pd.DataFrame(data)

    # Get all Pokemons whose spawn rate is less than 5%
    spawn_rate_less_than_5 = df[df["spawn_chance"] < 0.05]

    # Get all Pokemons that have less than 4 weaknesses
    less_than_4_weaknesses = df[df["weaknesses"].apply(lambda x: len(x) < 4)]

    # Get all Pokemons that have no multipliers at all
    no_multipliers = df[df["multipliers"].apply(lambda x: len(x) == 0)]

    # Get all Pokemons that do not have more than 2 evolutions
    less_than_2_evolutions = df[df["next_evolution"].apply(lambda x: len(x) <= 2)]

    # Get all Pokemons whose spawn time is less than 300 seconds
    df["spawn_time"] = pd.to_datetime(df["spawn_time"], format="%M:%S")
    spawn_time_less_than_300 = df[df["spawn_time"].dt.total_seconds() < 300]

    # Get all Pokemon who have more than two types of capabilities
    df["total_capabilities"] = df["type"].apply(lambda x: len(x))
    more_than_2_capabilities = df[df["total_capabilities"] > 2]

    return spawn_rate_less_than_5, less_than_4_weaknesses, no_multipliers, less_than_2_evolutions, spawn_time_less_than_300, more_than_2_capabilities

def plot_data(data):
    # Plot the distribution of spawn rate
    plt.hist(data["spawn_chance"], bins=10)
    plt.xlabel("Spawn Rate")
    plt.ylabel("Frequency")
    plt.title("Distribution of Spawn Rate")
    plt.show()

    # Plot the number of weaknesses for each Pokemon
    weakness_counts = data["weaknesses"].apply(len)
    plt.hist(weakness_counts, bins=range(8))
    plt.xlabel("Number of Weaknesses")
    plt.ylabel("Frequency")
    plt.title("Number of Weaknesses for Each Pokemon")
    plt.show()

    # Plot the number of evolutions for each Pokemon
    evolution_counts = data["next_evolution"].apply(len)
    plt.hist(evolution_counts, bins=range(5))
    plt.xlabel("Number of Evolutions")
    plt.ylabel("Frequency")
    plt.title("Number of Evolutions for Each Pokemon")
    plt.show()

if __name__ == "__main__":
    url = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"

    # Read the data from the URL into a dictionary
    data = pd.read_json(url)["pokemon"]

    # Analyze the data and answer the questions
    spawn_rate_less_than_5, less_than_4_weaknesses, no_multipliers, less_than_2_evolutions, spawn_time_less_than_300, more_than_2_capabilities = analyze_data(data)

    # Plot the analysis for better visualizations
    plot_data(spawn_rate_less_than_5)
    plot_data(less_than_4_weaknesses)
    plot_data(no_multipliers)
    plot_data(less_than_2_evolutions)
    plot_data(spawn_time_less_than_300)
    plot_data(more_than_2_capabilities)
```

● Get all the Earth meteorites that fell before the year 2000
● Get all the earth meteorites co-ordinates who fell before the year 1970
● Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more
than 10000kg

```python
import pandas as pd
import matplotlib.pyplot as plt

def analyze_data(data):
    # Create a DataFrame from the data
    df = pd.DataFrame(data)

    # Convert year column to datetime
    df['year'] = pd.to_datetime(df['year'], format='%Y')

    # Get all the Earth meteorites that fell before the year 2000
    meteorites_before_2000 = df[df['year'] < pd.Timestamp('2000-01-01')]

    # Get all the Earth meteorites coordinates that fell before the year 1970
    meteorites_before_1970_coords = df[df['year'] < pd.Timestamp('1970-01-01')][['reclat', 'reclong']]

    # Assuming mass is in kg, get all the Earth meteorites with mass > 10000kg
    mass_more_than_10000 = df[df['mass'] > 10000]

    return meteorites_before_2000, meteorites_before_1970_coords, mass_more_than_10000

def plot_data(data):
    # Plot the distribution of years for Earth meteorites falling before 2000
    plt.hist(data['year'].dt.year, bins=range(1800, 2000, 10))
    plt.xlabel('Year')
    plt.ylabel('Frequency')
    plt.title('Distribution of Earth Meteorites Falling Before 2000')
    plt.show()

    # Plot the coordinates of Earth meteorites falling before 1970
    plt.scatter(data['reclong'], data['reclat'], s=5)
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Coordinates of Earth Meteorites Falling Before 1970')
    plt.show()

if __name__ == "__main__":
    url = 'https://data.nasa.gov/resource/y77d-th95.json'

    # Read the data from the URL into a DataFrame
    data = pd.read_json(url)

    # Analyze the data and answer the questions
    meteorites_before_2000, meteorites_before_1970_coords, mass_more_than_10000 = analyze_data(data)

    # Plot the analysis for better visualizations
    plot_data(meteorites_before_2000)
    plot_data(meteorites_before_1970_coords)

    # Since mass is a float column, we can plot a histogram for mass_more_than_10000
    plt.hist(mass_more_than_10000['mass'], bins=10)
    plt.xlabel('Mass (kg)')
    plt.ylabel('Frequency')
    plt.title('Distribution of Earth Meteorites with Mass > 10000kg')
    plt.show()
```

## Qn.10.Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.
Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

def count_pos_tags(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Perform Part-of-Speech (POS) tagging
    tagged_words = pos_tag(words)

    # Initialize count variables
    verb_count = 0
    noun_count = 0
    pronoun_count = 0
    adjective_count = 0

    # Count the number of each POS tag
    for word, tag in tagged_words:
        if tag.startswith('VB'):  # Verbs
            verb_count += 1
        elif tag.startswith('NN'):  # Nouns
            noun_count += 1
        elif tag == 'PRP' or tag == 'PRP$':  # Pronouns
            pronoun_count += 1
        elif tag.startswith('JJ'):  # Adjectives
            adjective_count += 1

    # Create a dictionary with the counts
    pos_counts = {
        'Verbs': verb_count,
        'Nouns': noun_count,
        'Pronouns': pronoun_count,
        'Adjectives': adjective_count
    }

    return pos_counts

if __name__ == "__main__":
    # Example test cases
    text = "I am learning Python programming."
    pos_counts = count_pos_tags(text)
    print(pos_counts)

    text2 = "The cat sat on the mat."
    pos_counts2 = count_pos_tags(text2)
    print(pos_counts2)

    # Additional test cases
    text3 = "She quickly ran to catch the bus."
    pos_counts3 = count_pos_tags(text3)
    print(pos_counts3)

    text4 = "The big red apple fell from the tree."
    pos_counts4 = count_pos_tags(text4)
    print(pos_counts4)
```
